```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("merged-df.csv")
```

```python
data.head()
```

```python
# Convert the timestamp column to datetime format
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
# Set the timestamp as the index
data.set_index('timestamp', inplace=True)
```

```python
# Resample the data to daily frequency and calculate the mean for each day
daily_data = data.resample('D').mean()

# Display the first few rows of the resulting daily data
daily_data.head()
```

```python

# Reset the index and rename the timestamp column to 'date'
daily_data.reset_index(inplace=True)
daily_data['date'] = daily_data['timestamp'].dt.date
daily_data = daily_data.drop(columns=['timestamp'])  # Remove the old timestamp column
daily_data = daily_data.rename(columns={'date': 'date'})  # Rename to 'date'

# Display the first few rows of the daily data
daily_data.head()
```

```python
#bringing last column into 1st posiiton
cols = list(daily_data.columns)
cols = [cols[-1]] + cols[:-1]
daily_data = daily_data[cols]
```

```python
daily_data.head()
```

```python
#cancelling Unnamed column at position 1
daily_data = daily_data.drop(daily_data.columns[1], axis=1)
```

```python
daily_data.head()
```

```python
# Save to a new CSV file
daily_data.to_csv('daily_data.csv')
```
